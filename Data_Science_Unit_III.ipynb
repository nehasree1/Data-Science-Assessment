## Import Libraries

```
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import pandas as pd
```

## Basic Line Plot

```
x = [1,2,3,4]
y = [10,20,25,30]

plt.plot(x,y)
plt.title("Basic Line Plot")
plt.show()
```

## Subplots Example

```
x = np.array([0,1,2,3])
y1 = np.array([3,8,1,10])
y2 = np.array([10,20,30,40])

plt.subplot(1,2,1)
plt.plot(x,y1)

plt.subplot(1,2,2)
plt.plot(x,y2)

plt.show()
```

## Multiple Plots

```
fig, axs = plt.subplots(2,2)

axs[0,0].plot([1,2,3],[4,5,6])
axs[0,1].scatter([1,2,3],[4,5,6])
axs[1,0].bar([1,2,3],[4,5,6])
axs[1,1].hist([1,2,2,3,3,3,4,5])

plt.show()
```

## Controlling Axes

```
plt.plot([1,2,3,4],[10,20,25,30])
plt.xlim(0,5)
plt.ylim(0,35)
plt.show()
```

## Labels and Legend

```
plt.plot([1,2,3,4],[10,20,25,30], label='Line 1')
plt.plot([1,2,3,4],[30,25,20,15], label='Line 2')

plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Legend Example")
plt.legend()

plt.show()
```

## Annotation Example

```
x = [1,2,3,4,5]
y = [2,4,6,8,10]

plt.plot(x,y, marker='o')

plt.annotate("Point", xy=(4,8), xytext=(2,9),
             arrowprops=dict(facecolor='black'))

plt.show()
```

## Save Plot

```
plt.plot([1,2,3],[4,5,6])
plt.savefig("lineplot.png")
plt.show()
```

## Seaborn Scatter Plot

```
data = {"x":[1,2,3,4,5], "y":[2,4,3,5,1], "group":["X","Y","X","Y","X"]}
df = pd.DataFrame(data)

sns.scatterplot(x="x", y="y", hue="group", data=df)
plt.show()
```

## Joint Plot

```
data = {"x":[1,2,3,4,5], "y":[2,5,3,8,10]}
sns.jointplot(x="x", y="y", data=data)
plt.show()
```

## Pair Plot

```
data = {"x":[1,2,3,4,5], "y":[2,4,3,5,1], "z":[10,15,20,25,30]}
df = pd.DataFrame(data)

sns.pairplot(df)
plt.show()
```

## Histogram

```
data = np.random.randn(1000)

plt.hist(data)
plt.show()
```

## Box Plot

```
data = {"A":[1,2,3,4,5,30], "B":[2,4,6,8,7,28], "C":[3,6,9,5,2,7]}
df = pd.DataFrame(data)

sns.boxplot(data=df)
plt.show()
```

## Word Cloud

```
from wordcloud import WordCloud

text = "Data science visualization matplotlib seaborn python"
wc = WordCloud().generate(text)

plt.imshow(wc)
plt.axis("off")
plt.show()
```

## 3D Scatter Plot

```
from mpl_toolkits.mplot3d import Axes3D

x = [1,2,3,4,5]
y = [2,4,3,5,1]
z = [10,15,20,25,30]

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.scatter(x,y,z)

plt.show()
```
